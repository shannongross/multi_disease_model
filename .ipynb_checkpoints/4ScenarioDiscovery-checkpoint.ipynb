## PROBLEM FORMULATION 1 
#### Retrieve robust results

```python
nfes = 15000
for PF in [1]:
    print('Retrieving robust candidates under Problem Formulation {} '.format(PF) )
    file_name = r'.\results\robust_candidates\robust_PF{}_nfes{}'.format(PF,nfes) + '.tar.gz'
    pareto_results = load_results(file_name)
    experiments, outcomes = pareto_results
    disease_model = get_model_for_problem_formulation(PF)   
    cleaned_exp = experiments.drop(labels=[l.name for l in disease_model.levers], axis=1)
    cleaned_exp = cleaned_exp.drop([ 'policy','Unnamed: 0'], axis=1)
```

Discover **Mortality** vulnerabilities

```python
ooi = 'Mortality'

x = cleaned_exp #independent variable experiments
y = outcomes[ooi] > np.percentile(outcomes[ooi], 90) 
    #focus on 10-20 percent of the worst outcomes for this objective 
prim_alg = prim.Prim(x, y, threshold=0.8, peel_alpha=0.1)
box_mort = prim_alg.find_box()

box_mort.show_tradeoff()
plt.title("Peeling Trajectory for *{}* under PF {}".format(ooi,PF))
plt.figure(figsize=(4,4)) 
plt.show()
```

```python
boi=12
box_mort.inspect(boi, style='graph')
print(prim_alg.stats_to_dataframe())
# box1.show_ppt()
```

Discover **Morbidity** vulnerabilities

```python
ooi = 'Morbidity'

data = outcomes[ooi]    
y = data > np.percentile(data, 75) 
    #focus on 10-20 percent of the worst outcomes for this objective   
prim_alg = prim.Prim(cleaned_exp, y, threshold=0.65, peel_alpha=.1)
box_morb = prim_alg.find_box()  

box_morb.show_tradeoff()
plt.title("Peeling Trajectory for *{}* under PF {}".format(ooi,PF))
plt.figure(figsize=(4,4)) 
plt.show()
```

```python
boi=9

box_morb.inspect(boi, style='graph')
print(prim_alg.stats_to_dataframe())
```
